```python
def visualize_show(save_f,**images):
    """PLot images in one row."""
    n = len(images)
    plt.figure(figsize=(16, 5))
    for i, (name, image) in enumerate(images.items()):
        plt.subplot(1, n, i + 1)
        plt.xticks([])
        plt.yticks([])
        plt.title(' '.join(name.split('_')).title())
        plt.imshow(image)
    plt.show()
    #plt.savefig(save_f)
    plt.close()
```

```python
def visualize_overlap(save_f,mask0,mask2,**images):
    """PLot images in one row."""
    n = len(images)
    plt.figure(figsize=(16, 5))
    mask.append(mask0)
    mask.append(mask2)
    for i, (name, image) in enumerate(images.items()):
        plt.subplot(1, n, i + 1)
        plt.xticks([])
        plt.yticks([])
        plt.title(' '.join(name.split('_')).title())
        plt.imshow(image)
        plt.imshow(mask[i], cmap='jet', alpha=0.5)
        
        
    plt.show()
    #plt.savefig(save_f)
    plt.close()
```

```python
name = 'XJTr0228'
bone_mip0 = np.load(bone_mip0_path)['bone_mip0']
bone_mip1 = np.load(bone_mip1_path)['bone_mip1']
bone_mip2 = np.load(bone_mip2_path)['bone_mip2']
print (bone_mip1.shape)
visualize_show(name+'.png',bone_mip0=bone_mip0,bone_mip1=bone_mip1,bone_mip2=bone_mip2,)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = torch.load('./model/epoch190_model.pth')
model.eval()
```

```python

shape1 = bone_mip1.shape
shape2 = bone_mip2.shape

image_1 = cv2.resize(bone_mip1, dsize= (512, 512), interpolation=cv2.INTER_NEAREST)
image_2 = cv2.resize(bone_mip1, dsize= (512, 512), interpolation=cv2.INTER_NEAREST)

image_1 = torch.FloatTensor(np.tile(image_1.copy(), (3, 1, 1)))
image_2 = torch.FloatTensor(np.tile(image_2.copy(), (3, 1, 1))) 

```

```python
image_1 = image_1.unsqueeze(0).to(device)
image_2 = image_2.unsqueeze(0).to(device)

with torch.no_grad():
    predict_1 = model(image_1)
    predict_2 = model(image_2)
    
```

```python
rows1, cols1 = [int(s) for s in shape1]
rows2, cols2 = [int(s) for s in shape2]
```

```python
pre_mask_1 = torch.argmax(predict_1.squeeze(), 0)
pre_mask_2 = torch.argmax(predict_2.squeeze(), 0)

pre_mask_1 = pre_mask_1 >= 0.5
output_pre_mask_1 = pre_mask_1.cpu().numpy().astype(int)

pre_mask_2 = pre_mask_2 >= 0.5
output_pre_mask_2 = pre_mask_2.cpu().numpy().astype(int)

visualize_show(name + '.png', output_pre_mask_1=output_pre_mask_1, output_pre_mask_2=output_pre_mask_2)
```

```python
resized_mask_1 = cv2.resize(output_pre_mask_1, dsize=(cols1, rows1), interpolation=cv2.INTER_NEAREST)
resized_mask_2 = cv2.resize(output_pre_mask_2, dsize=(cols2, rows2), interpolation=cv2.INTER_NEAREST)

visualize_show(name + '.png', resized_mask_1=resized_mask_1, resized_mask_2=resized_mask_2)
```

```python
one_region_refined_1, largest_removed, kept_size = remove_all_but_the_largest_connected_component(resized_mask_1, [1])
one_region_refined_2, largest_removed, kept_size = remove_all_but_the_largest_connected_component(resized_mask_2, [1])
visualize_show(os.path.join(brain_mip_2d_save, name + '_befordialation.png'),
          one_region_refined_1=one_region_refined_1, one_region_refined_2=one_region_refined_2, )
```

```python
# dilate
dilate_refined_1 = dilation(one_region_refined_1, selem=disk(20))
dilate_refined_2 = dilation(one_region_refined_2, selem=disk(20))

visualize_show(name + '.png', dilate_refined_1=dilate_refined_1, dilate_refined_2=dilate_refined_2)
```

```python
image_1 = cv2.resize(image_1.cpu().numpy().squeeze()[0], dsize=(cols1, rows1), interpolation=cv2.INTER_NEAREST)
image_2 = cv2.resize(image_2.cpu().numpy().squeeze()[0], dsize=(cols1, rows1), interpolation=cv2.INTER_NEAREST)

visualize_overlap(save_f = os.path.join(brain_mip_2d_save, name + '.png'),mask0=dilate_refined_1, mask1=dilate_refined_2, image_1=image_1,
                  image_2=image_2, )
```
