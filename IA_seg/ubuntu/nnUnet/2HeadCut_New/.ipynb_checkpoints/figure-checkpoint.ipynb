```python
def visualize_show(save_f,**images):
    """PLot images in one row."""
    n = len(images)
    plt.figure(figsize=(16, 5))
    for i, (name, image) in enumerate(images.items()):
        plt.subplot(1, n, i + 1)
        plt.xticks([])
        plt.yticks([])
        plt.title(' '.join(name.split('_')).title())
        plt.imshow(image)
    plt.show()
    plt.savefig(save_f)
    plt.close()
```

```python
def visualize_overlap(save_f,mask0,mask2,**images):
    """PLot images in one row."""
    n = len(images)
    plt.figure(figsize=(16, 5))
    mask.append(mask0)
    mask.append(mask2)
    for i, (name, image) in enumerate(images.items()):
        plt.subplot(1, n, i + 1)
        plt.xticks([])
        plt.yticks([])
        plt.title(' '.join(name.split('_')).title())
        plt.imshow(image)
        plt.imshow(mask[i], cmap='jet', alpha=0.5)
        
        
    plt.show()
    plt.savefig(save_f)
    plt.close()
```

現成的bone2D image作為輸入

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = torch.load('./model/epoch190_model.pth')
model.eval()
```

```python
shape1 = bone_mip1.shape
shape2 = bone_mip2.shape

image_1 = cv2.resize(bone_mip1, dsize= (512, 512), interpolation=cv2.INTER_NEAREST)
image_2 = cv2.resize(bone_mip2, dsize= (512, 512), interpolation=cv2.INTER_NEAREST)

image_1 = torch.FloatTensor(np.tile(image_1.copy(), (3, 1, 1)))
image_2 = torch.FloatTensor(np.tile(image_2.copy(), (3, 1, 1))) 

```

```python
image_1 = image_1.unsqueeze(0).to(device)
image_2 = image_2.unsqueeze(0).to(device)

with torch.no_grad():
    predict_1 = model(image_1)
    predict_2 = model(image_2)
    
```

```python
rows1, cols1 = [int(s) for s in shape1]
rows2, cols2 = [int(s) for s in shape2]
```

```python
pre_mask_1 = torch.argmax(predict_1.squeeze(), 0)
pre_mask_2 = torch.argmax(predict_2.squeeze(), 0)

pre_mask_1 = pre_mask_1 >= 0.5
output_pre_mask_1 = pre_mask_1.cpu().numpy().astype(int)

pre_mask_2 = pre_mask_2 >= 0.5
output_pre_mask_2 = pre_mask_2.cpu().numpy().astype(int)

visualize_show(name + '.png', output_pre_mask_1=output_pre_mask_1, output_pre_mask_2=output_pre_mask_2)
```

```python
resized_mask_1 = cv2.resize(output_pre_mask_1, dsize=(cols1, rows1), interpolation=cv2.INTER_NEAREST)
resized_mask_2 = cv2.resize(output_pre_mask_2, dsize=(cols2, rows2), interpolation=cv2.INTER_NEAREST)

visualize_show(name + '.png', resized_mask_1=resized_mask_1, resized_mask_2=resized_mask_2)
```
