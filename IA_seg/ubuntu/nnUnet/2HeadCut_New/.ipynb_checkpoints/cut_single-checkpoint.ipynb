```python
#ori_folder = 'F:/data/CellPress1338'
dst_folder = 'F:/data/CellPress1338/headcut_all'
name = 'Ts0001'
```

```python
def load_pickle(file: str, mode: str = 'rb'):
    with open(file, mode) as f:
        a = pickle.load(f)
    return a
```

```python
just_head_properties_save = './after_headcut_properties_CellPressAll.pkl'
all_properties = load_pickle(just_head_properties_save)
```

```python
def save_pickle(obj, file: str, mode: str = 'wb') -> None:
    with open(file, mode) as f:
        pickle.dump(obj, f)
```

```python
def maybe_create_path(path):
    if not os.path.exists(path):
        os.makedirs(path)
    return path
```

```python
def save_itk_from_numpy(numpy_data, property):

    pred_itk_image = sitk.GetImageFromArray(numpy_data)
    pred_itk_image.SetSpacing(property["itk_spacing"])
    pred_itk_image.SetOrigin(property["itk_origin"])

    return pred_itk_image
```

```python
def get_cube_as_coords(image,min_z,max_z,min_y,max_y,min_x,max_x):
    cube_image = image[min_z:max_z,min_y:max_y,min_x:max_x]
    return cube_image
```

```python
def visualize_show(save_f,**images):
    """PLot images in one row."""
    n = len(images)
    plt.figure(figsize=(16, 5))
    for i, (name, image) in enumerate(images.items()):
        plt.subplot(1, n, i + 1)
        plt.xticks([])
        plt.yticks([])
        plt.title(' '.join(name.split('_')).title())
        plt.imshow(image)
    plt.show()
    #plt.savefig(save_f)
    plt.close()
```

現成的bone2D image作為輸入

```python
min_z, max_z, min_y, max_y, min_x, max_x = 365,693,108,449,118,382
```

```python
all_properties[name]['brain_coords'] = min_z, max_z, min_y, max_y, min_x, max_x
print ('after dialation:',min_z, max_z, min_y, max_y, min_x, max_x)
```

```python
cta_nii = sitk.ReadImage(all_properties[name]['ori_cta'])
cta_np = sitk.GetArrayFromImage(cta_nii).astype(np.float32)
seg_nii = sitk.ReadImage(all_properties[name]['ori_seg'])
seg_np = sitk.GetArrayFromImage(seg_nii).astype(np.int32)

brain_cta = get_cube_as_coords(cta_np,min_z, max_z, min_y, max_y, min_x, max_x)
brain_seg = get_cube_as_coords(seg_np, min_z, max_z, min_y, max_y, min_x, max_x)

print ('before shape:',cta_np.shape)
print ('after shape:',brain_cta.shape)
```

```python
just_head_cta_save = os.path.join(dst_folder,'cta_img')
just_head_seg_save = os.path.join(dst_folder,'ane_seg')
```

```python
brain_cta_nii = save_itk_from_numpy(brain_cta, all_properties[name])
brain_seg_nii = save_itk_from_numpy(brain_seg, all_properties[name])
    
all_properties[name]['before_size'] = cta_np.shape
all_properties[name]['after_size'] = brain_cta.shape
all_properties[name]['just_head_cta_save'] = os.path.join(just_head_cta_save,name+'.nii.gz')
all_properties[name]['just_head_seg_save'] = os.path.join(just_head_seg_save, name + '.nii.gz')

sitk.WriteImage(brain_cta_nii, all_properties[name]['just_head_cta_save'])
sitk.WriteImage(brain_seg_nii, all_properties[name]['just_head_seg_save'])

visualize_show(os.path.join(name+'.png'),x=brain_cta[:,:,brain_cta.shape[2]//2],
z=brain_cta[brain_cta.shape[0]//2,:,:],y=brain_cta[:,brain_cta.shape[1]//2,:])

save_pickle(all_properties, just_head_properties_save)
```
