```python
from collections import OrderedDict
import pprint, pickle
```

```python
pkl_file = open('after_headcut_properties_newname.pkl', 'rb')
data1 = pickle.load(pkl_file)
pprint.pprint(len(data1))
pprint.pprint(data1)
pkl_file.close()
```

```python
XJTs_list = []
for i in range(566):
    num = str(i).zfill(4)
    new_name = 'XJTsN'+num
    XJTs_list.append(new_name)
print (len(XJTs_list))
print (XJTs_list[-5:])
```

```python
properti = OrderedDict()
i=0
for item in data1:
    properti[XJTs_list[i]] = data1[item]
    i+=1
    
    
```

```python
import pickle
def save_pickle(obj, file: str, mode: str = 'wb') -> None:
    with open(file, mode) as f:
        pickle.dump(obj, f)
```
