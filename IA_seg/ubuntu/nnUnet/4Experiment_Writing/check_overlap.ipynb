```python
aneurysm_path = "./xj_list.txt"
with open(aneurysm_path,encoding='utf-8') as f:
    class_names = f.readlines()# 全部读取最后返回一个列表存所有的类,每行后面都会带有“\n”
    aneurysm_path = [c.strip() for c in class_names]# 采用strip（）方法可以去掉最后的“\n”
    print (type(aneurysm_path))
    print (len(aneurysm_path))
    f.close()
print (aneurysm_path)
rename_dict = []
for line in aneurysm_path:
    origi_name = line.split('\t')[0]
    dst_name = line.split('\t')[1]
    age = line.split('\t')[2]
    rename_dict.append([origi_name,dst_name,age])
print (rename_dict)
```

```python
xj_list = rename_dict
```

```python
cellpress110 = rename_dict
```

```python
for i in xj_list:
    slic,voxel,gender = i[0],i[1],i[2]
    for j in cellpress110:
        slicj,voxelj,genderj = j[0],j[1],j[2]
        if slic==slicj:
            if voxel==voxelj and gender==genderj:
                print ('xj_list',i,'has same slice and voxels gender')
            else:
                continue
        else:
            continue
            
```
