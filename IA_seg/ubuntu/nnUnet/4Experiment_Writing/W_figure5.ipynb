```python
def read_itk(path):
    data = sitk.ReadImage(path)
    spacing = data.GetSpacing()
    scan = sitk.GetArrayFromImage(data)
    return scan
```

```python
def mid_slice(List):
    new_list = []

    def is_continues(value, List, contin, new_list,new_idx):

        if len(List) == 1:
            new_list.append(value)
            return new_list

        if value == (List[0] - 1):
            try:
                new_list[new_idx] = value
            except:
                new_list.append(value)
            contin += 1
            return is_continues(List[0], List[1:], contin, new_list,new_idx)
        else:
            try:
                new_list[new_idx] = value - contin // 2
                new_idx += 1
            except:
                new_list.append(value - contin // 2)
                new_idx +=1
            contin = 0
            return is_continues(List[0], List[1:], contin, new_list,new_idx)

    return is_continues(List[0], List[1:], 0, new_list,0)
```

```python
def plot_inference(scan,cta,inference_dir):
    """
    visualize 'Sagittal'
    """
    
    img_list = []
    for img in scan:
        iD = os.path.basename(img)[:-7]
        img_list.append([iD,read_itk(img)])
        
    cta_img = read_itk(cta)
    
    num_slices = img_list[0][1].shape[0]
 

    plot_s = []
    for s in range(num_slices):
        if any([img_list[0][1][s,:,:].sum()!=0,img_list[-1][1][s,:,:].sum()!=0]):
            plot_s.append(s)
    print (plot_s)
    plot_s_mid = mid_slice(plot_s)
    print ('plot_s_mid:',plot_s_mid)
    num_row = len(plot_s_mid)
    num_column = len(scan)
    
    f, plots = plt.subplots(num_row, num_column, figsize=(num_column*5, num_row*5))
    
    for i,ps in enumerate(plot_s_mid):
        for j in range(num_column):
            plot = plots[i,j]
            plot.axis('off')
            plot.imshow(cta_img[ps,:,:])
            plot.imshow(img_list[j][1][ps,:,:],cmap='jet', alpha=0.5)
            if img_list[j][1][:][:][ps].sum()!=0:
                flag = 'pos'
            else:
                flag = 'neg'
            plot.set_title('slice_{}_{}'.format(img_list[j][0],flag))
        
    f.savefig(os.path.join(inference_dir,iD + 'visual.png'))
        
```

```python
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import numpy as np

def plot_3d(save_filename,scan,head,threshold=100):
    
    # Position the scan upright, 
    # so the head of the patient would be at the top facing the camera
    
    
    #mask
    iD = os.path.basename(scan)[:-7]
    image = read_itk(scan)

    img = image.transpose(2,1,0)
    position = np.nonzero(img)

    fig = plt.figure(figsize=(10, 10))
    
    ax = fig.add_subplot(111, projection='3d')
    
    mesh = ax.scatter(position[0],position[1],position[2])
    #face_color = [1,0,1]
    face_color = [0,1,1]
    
    mesh.set_facecolor(face_color)
    ax.add_collection3d(mesh)
    
    
    # head
    head_np = read_itk(head)
    head_img = head_np.transpose(2,1,0)
    head_img = np.clip(head_img, 0, 800)
    verts, faces, _, x = measure.marching_cubes_lewiner(head_img, threshold)
    head_mesh = Poly3DCollection(verts[faces], alpha=0.1)
    head_color = [0, 0, 0]
    head_mesh.set_facecolor(head_color)
    ax.add_collection3d(head_mesh)
    
    

    ax.set_xlim(0, img.shape[0])
    ax.set_ylim(0, img.shape[1])
    ax.set_zlim(0, img.shape[2])
    ax.set_title(iD)
    
    #save_filename = os.path.join(os.path.dirname(scan),iD + '_3D.png')
    
    ax.axis('off')
    plt.savefig(save_filename,dpi=500,bbox_inches = 'tight')
    #plt.show()
    plt.close()
    return save_filename
    
```
