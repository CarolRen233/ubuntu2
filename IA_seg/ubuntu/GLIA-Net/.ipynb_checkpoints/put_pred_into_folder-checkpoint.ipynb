```python
import os,glob
import shutil
```

```python
ori_folder = '/home/ubuntu/codes/MyMedicalZoo/output/XJTsN'
```

```python
def maybe_create_path(path):
    if not os.path.exists(path):
        os.makedirs(path)
    return path
```

```python
maybe_create_path(os.path.join(ori_folder,'_headcut_prob'))
maybe_create_path(os.path.join(ori_folder,'_headcut_pred'))
maybe_create_path(os.path.join(ori_folder,'_headcut_refined'))
```

```python
prob_list = sorted(glob.glob(os.path.join(ori_folder,'*_headcut__prob.nii.gz')))
pred_list = sorted(glob.glob(os.path.join(ori_folder,'*_headcut__pred.nii.gz')))
refined_list = sorted(glob.glob(os.path.join(ori_folder,'*_headcut__refined.nii.gz')))
print (len(prob_list))
```

```python
def put_into_folder(ori_list,save_folder):
    for ori_file in ori_list:
        name = os.path.basename(ori_file)
        shutil.move(ori_file,os.path.join(save_folder,name))
        
```

```python
put_into_folder(prob_list,os.path.join(ori_folder,'_headcut_prob'))
put_into_folder(pred_list,os.path.join(ori_folder,'_headcut_pred'))
put_into_folder(refined_list,os.path.join(ori_folder,'_headcut_refined'))
```

```python
prob_list = sorted(glob.glob(os.path.join(ori_folder,'_headcut_prob','*_headcut__prob.nii.gz')))
pred_list = sorted(glob.glob(os.path.join(ori_folder,'_headcut_pred','*_headcut__pred.nii.gz')))
refined_list = sorted(glob.glob(os.path.join(ori_folder,'_headcut_refined','*_headcut__refined.nii.gz')))
print (len(prob_list))
```

```python
def rename_file(lista,suffix):
    for item in lista:
        print (item)
        name = os.path.basename(item).split(suffix)[0]
        print (name)
        os.rename(item,os.path.join(os.path.dirname(item),name+'.nii.gz'))
```

```python
rename_file(prob_list,'_headcut__prob')
rename_file(pred_list,'_headcut__pred')
rename_file(refined_list,'_headcut__refined')
```
