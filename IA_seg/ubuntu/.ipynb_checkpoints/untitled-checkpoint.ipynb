```python
import os,glob
import shutil
```

```python
ori_test_path = '/mnt/f/data/CellPress1338/cta_img'
dst_path = '/mnt/f/data/CellPress1338/cta_img_ts'
```

```python
XJTr_list = []
for i in range(500):
    num = str(i).zfill(4)
    new_name = 'XJTr'+num
    XJTr_list.append(new_name)
print (len(XJTr_list))
print (XJTr_list[-5:])


XJTs_list = []
for i in range(66):
    num = str(i).zfill(4)
    new_name = 'XJTs'+num
    XJTs_list.append(new_name)
print (len(XJTs_list))
print (XJTs_list[-5:])
```

```python
for i,name in enumerate(XJTr_list):
    
    new_name = 'XJTsN' + str(i+66).zfill(4)
    
    print (name)
    print (new_name)
    
            
    ori_cta_file = os.path.join(ori_test_path,'imagesTr',name)
    ori_seg_file = os.path.join(ori_test_path,'labelsTr',name+'.nii.gz')
    
    dst_cta_file = os.path.join(dst_path,'imagesTs',new_name)
    dst_seg_file = os.path.join(dst_path,'labelsTs',new_name+'.nii.gz')
    
    os.symlink(ori_cta_file+'_0000.nii.gz', dst_cta_file+'_0000.nii.gz')
    os.symlink(ori_cta_file+'_0001.nii.gz', dst_cta_file+'_0001.nii.gz')
    os.symlink(ori_cta_file+'_0002.nii.gz', dst_cta_file+'_0002.nii.gz')
    os.symlink(ori_seg_file, dst_seg_file)

```

```python
for i,name in enumerate(XJTs_list):
    
    new_name = 'XJTsN' + str(i).zfill(4)
    
    
    print (name)
    print (new_name)
    
            
    ori_cta_file = os.path.join(ori_test_path,'imagesTs',name)
    ori_seg_file = os.path.join(ori_test_path,'labelsTs',name+'.nii.gz')
    
    dst_cta_file = os.path.join(dst_path,'imagesTs',new_name)
    dst_seg_file = os.path.join(dst_path,'labelsTs',new_name+'.nii.gz')
    
    os.symlink(ori_cta_file+'_0000.nii.gz', dst_cta_file+'_0000.nii.gz')
    os.symlink(ori_cta_file+'_0001.nii.gz', dst_cta_file+'_0001.nii.gz')
    os.symlink(ori_cta_file+'_0002.nii.gz', dst_cta_file+'_0002.nii.gz')
    os.symlink(ori_seg_file, dst_seg_file)

```
